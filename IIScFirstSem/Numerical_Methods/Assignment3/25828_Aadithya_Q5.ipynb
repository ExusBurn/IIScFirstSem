```python
#Obtaining function value
import math
import matplotlib.pyplot as plt
def f(x):
    return math.exp(-x**2)
I_Exact = math.sqrt(math.pi)

#Now, the trapezoidal rule(With Summation to make it composite)
def trapezoidal_rule(f, a, n):
    # Integration limits
    x_start = -a
    x_end = a
    
    # Step size
    h = (x_end - x_start) / n
    
    # Initialize sum
    total = 0.5 * (f(x_start) + f(x_end))
    
    # Loop through internal points
    for i in range(1, n):
        x_i = x_start + i * h
        total += f(x_i)
    
    # Multiply by step size
    I = h * total
    return I

#Now, calling it for the values mentioned
a_values = [0.5, 1, 2, 5]
n_values = [10, 50, 500, 3000]

# Store results for table
results = []

for a in a_values:
    for n in n_values:
        I_approx = trapezoidal_rule(f, a, n)
        print(f"a={a}, n={n}, I≈{I_approx:.8f}")
print(f"Exact Value:={I_Exact}")
```

```python
#Part (c), obtaining the absolute error
# Store results for table
results = []

for a in a_values:
    for n in n_values:
        I_approx = trapezoidal_rule(f, a, n)
        abs_error = abs(I_Exact - I_approx)
        results.append((a, n, I_approx, abs_error))

#Part (d) - Print the table
print(f"{'a':>5} {'n':>5} {'I_approx':>15} {'Abs Error':>15}")
print("-"*50)
for row in results:
    a, n, I_approx, abs_error = row
    print(f"{a:5.1f} {n:5d} {I_approx:15.6e} {abs_error:15.6e}")

# Part (e) - Plot absolute error vs a for large n
n_large = 3000
errors_for_large_n = []
a_grid = [0.5, 1, 2, 5]

for a in a_grid:
    I_approx = trapezoidal_rule(f, a, n_large)
    errors_for_large_n.append(abs(I_Exact - I_approx))

plt.figure(figsize=(6,4))
plt.plot(a_grid, errors_for_large_n, 'o-', color='blue')
plt.xlabel("a")
plt.ylabel("Absolute Error")
plt.title(f"Absolute Error vs a (n={n_large})")
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
