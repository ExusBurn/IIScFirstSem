# Bisection Method

```python
import numpy as np
import matplotlib as plt

def my_polynomial(x):
    return x**10 - 1
j=0
lower = 0
upper = 1.1
epsilon = 0.0001
while(upper-lower>epsilon):
    j+=1
    x = (upper + lower)/2
    value = my_polynomial(x)
    if value == 0:
        print(f"The final value is:={x}, number of iterations:={j} and perfect convergence was obtained")
        break

    if value*my_polynomial(lower)>0:
        lower = x
    else:
        upper = x
    if upper - lower < epsilon:
        print(f"The final value is:={x}, number of iterations:={j} and the error has been minimized")
    
    
    


```

# Modified Newton's Method and Newton's method(Part a)

```python
#First, need to obtain the 3 functions
import math
def my_polynomial_f(x):
  inner_term = x**2 - math.exp(-x) * math.cos(x)
  f_x = inner_term**2
  return f_x

def my_polynomial_f_prime(x):
  f_prime_part1 = 2 * (x**2 - math.exp(-x) * math.cos(x))
  f_prime_part2 = 2 * x + math.exp(-x) * (math.cos(x) + math.sin(x))
  f_prime_x = f_prime_part1 * f_prime_part2
  return f_prime_x

def my_polynomial_f_double_prime(x):
  f_prime_part2 = 2 * x + math.exp(-x) * (math.cos(x) + math.sin(x))
  f_double_prime_term1 = 2 * (f_prime_part2)**2
  f_double_prime_term2 = 2 * (x**2 - math.exp(-x) * math.cos(x)) * (2 - 2 * math.exp(-x) * math.sin(x))
  f_double_prime_x = f_double_prime_term1 + f_double_prime_term2
  return f_double_prime_x
print(my_polynomial_f(5))
```

# Modified Newton's Method and Newton's Method part (a)

```python
x_value = 5.0
j = 0
maxiter = 500
epsilon = 10**-7
#Newton's
import matplotlib.pyplot as plt
error_list = []
iteration_list = []
while j < maxiter:
    j+=1
    func_val = my_polynomial_f(x_value)
    diffy = my_polynomial_f_prime(x_value)
    if diffy == 0:
        print("Derivative is zero, can't continue")
        break
    newpoint = x_value - func_val/diffy
    if newpoint == 0:
        break
    error = abs((newpoint-x_value)/newpoint)
    error_list.append(error)
    iteration_list.append(j)
    if error < epsilon:
        print(f"Target hit! Value is {newpoint}, on iteration {j} and error is {error}")
        break
    x_value = newpoint
    print(x_value,error)
plt.figure(figsize=(10,6))
plt.semilogy(iteration_list,error_list,marker='o',linestyle='-')
plt.title("Newton's Method Rate of Convergence")
plt.xlabel("Iteration")
plt.ylabel("Relative Error(log scale)")
plt.grid(True,which="both",ls="--")
plt.show()




```

```python
#Modified Newton's
x_value = 5.0
j = 0
iteration_list = []
error_list = []
while j < maxiter:
    j+=1
    poly = my_polynomial_f(x_value)
    diff = my_polynomial_f_prime(x_value)
    doublediff = my_polynomial_f_double_prime(x_value)
    mu = poly/diff
    mu_diff = (diff*diff-poly*doublediff)/(diff*diff)
    func_val = mu
    diffy = mu_diff
    if diffy == 0:
        print("Derivative is zero, can't continue")
        break
    newpoint = x_value - func_val/diffy
    error = abs((newpoint-x_value)/newpoint)
    error_list.append(error)
    iteration_list.append(j)
    if error < epsilon:
        print(f"Target hit! Value is {newpoint}, on iteration {j} and error is {error}")
        break
    x_value = newpoint
    print(x_value,error)
plt.figure(figsize=(10,6))
plt.semilogy(iteration_list,error_list,marker='o',linestyle='-')
plt.title("Newton's Method Rate of Convergence")
plt.xlabel("Iteration")
plt.ylabel("Relative Error(log scale)")
plt.grid(True,which="both",ls="--")
plt.show()
```
