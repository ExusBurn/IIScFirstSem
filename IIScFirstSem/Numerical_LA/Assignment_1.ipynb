```python
    #First lets import some libraries
import numpy as np
A = np.random.randn(100,2)
#Now, the goal is to check for multiple random vectors x, multiply that by A
#And obtain the maximum value of (Axi^p)^1/p.
j = 0
store_norms_1 = []
x_val = [[]]
store_norms_2 = []
store_norms_3 = []
store_norms_4 = []
store_norms_5 = []
store_norms_6 = []
store_norms_inf = []
#First, let's do p = 1
while(j< 1000):
    j+=1
    temp = np.random.randn(2,1)
    temp = temp.squeeze()
    x = temp/np.linalg.norm(temp, ord=1) #Normalizing the vector gives us a random unit vector x make sure it is the pth norm!!!
    norm_Ax_1 = np.linalg.norm((A @ x).squeeze(), ord=1)
    store_norms_1.append(norm_Ax_1)
    x = temp/np.linalg.norm(temp, ord=2)
    norm_Ax_2 = np.linalg.norm((A @ x).squeeze(), ord=2)
    store_norms_2.append(norm_Ax_2)

    x = temp/np.linalg.norm(temp, ord=3)
    norm_Ax_3 = np.linalg.norm((A @ x).squeeze(), ord=3)
    store_norms_3.append(norm_Ax_3)
    x = temp/np.linalg.norm(temp, ord=4)
    norm_Ax_4 = np.linalg.norm((A @ x).squeeze(), ord=4)
    store_norms_4.append(norm_Ax_4)
    x = temp/np.linalg.norm(temp, ord=5)
    norm_Ax_5 = np.linalg.norm((A @ x).squeeze(), ord=5)
    store_norms_5.append(norm_Ax_5)
    x = temp/np.linalg.norm(temp, ord=6)
    norm_Ax_6 = np.linalg.norm((A @ x).squeeze(), ord=6)
    store_norms_6.append(norm_Ax_6)
    x = temp/np.linalg.norm(temp, ord=np.inf)
    norm_Ax_inf = np.linalg.norm((A @ x), ord=np.inf)
    store_norms_inf.append(norm_Ax_inf)
max_1_sim = max(store_norms_1)
max_2_sim = max(store_norms_2)
max_3_sim = max(store_norms_3) 
max_4_sim = max(store_norms_4) 
max_5_sim = max(store_norms_5) 
max_6_sim = max(store_norms_6)  
max_inf_sim = max(store_norms_inf)
max_1_actual = np.linalg.norm(A.squeeze(), ord=1)
max_2_actual = np.linalg.norm(A.squeeze(), ord=2)
max_inf_actual = np.linalg.norm(A.squeeze(), ord=np.inf)
print(f"Actual 1 norm:= {max_1_actual}, Simulation:= {max_1_sim}")
print(f"Actual 2 norm:= {max_2_actual}, Simulation:= {max_2_sim}")
print(f"Simulation of 3 norm:= {max_3_sim}")
print(f"Simulation of 4 norm:= {max_4_sim}")
print(f"Simulation of 5 norm:= {max_5_sim}")
print(f"Simulation of 6 norm:= {max_6_sim}")
print(f"Actual inf norm:= {max_inf_actual}, Simulation:= {max_inf_sim}")
```
